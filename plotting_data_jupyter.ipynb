```python
# Creating experiments dataframe
total = len(df_experiments)
completed = 0
for val in df_experiments["Experiment Complete"]:
    if val == 1:
        completed += 1
metadata = 0
for val in df_experiments["Metadata Complete"]:
    if val == 1:
        metadata += 1
validated = 0
for val in df_experiments["Experiment Validated"]:
    if val == 1:
        validated += 1
experiments = ("Total", "Completed", "Metadata", "Validated")
values = [total, completed, metadata, validated]
```

```python
# Creating experiments_prev dataframe
total = len(df_experiments_prev)
completed = 0
for val in df_experiments_prev["Experiment Complete"]:
    if val == 1:
        completed += 1
metadata = 0
for val in df_experiments_prev["Metadata Complete"]:
    if val == 1:
        metadata += 1
validated = 0
for val in df_experiments_prev["Experiment Validated"]:
    if val == 1:
        validated += 1
experiments2 = ("Total", "Completed", "Metadata", "Validated")
values2 = [total, completed, metadata, validated]
```

```python
# Inputting values for x and y
plt.bar(experiments, values)
# # Labeling Columns
plt.xlabel("Experiments")
plt.ylabel("Quantity")
plt.title("Experiment Data")
# # Show Graph
plt.show()
```

```python
# Inputting values for x and y
plt.bar(experiments2, values2)
# # Labeling Columns
plt.xlabel("Experiments")
plt.ylabel("Quantity")
plt.title("Previous Experiment Data")
# # Show Graph
plt.show()
```

```python
dc = {"Experiments": experiments, "Counts": values}
df = pd.DataFrame(dc, index=["Total", "Completed", "Metadata", "Validated"])
# print(df)
df.plot.pie(y='Counts', title="Experiment Totals", legend=False, \
                    autopct='%1.1f%%', explode=(0, 0, 0, 0), \
                    shadow=True, startangle=0) #(kind='pie', y="Percentages", figsize=(5, 5))
plt.show()
```
